```python
list_line = []
with open('179.csv', 'r') as f:
    for line in f:
        l = line.rstrip() 
        list_line.append(l.split(','))

```

```python
list_id = []
for line in list_line:
    list_id.append(line[1])
list_id.pop(0)

```

```python
list_coord = []

for line in list_line:
    list_coord.append([line[0],line[1],line[3],line[4]])
list_coord


```

```python
new_l = sorted(list_coord,key=lambda x: (x[0],x[1])) # use lambda to sort by "x[0]"-> first element of the sublists or x[1] -> second element, if its a tie
new_l[0]
```

```python
max_dist = 0
x_p = 0
y_p = 0
for t in set_id:
    dist = 0
    count = 0
    for i in range(0,len(new_l)):
        if count == 0 and new_l[i][1] == t:
            x_p = new_l[i][2]
            y_p = new_l[i][3]
            count+=1
        elif new_l[i][1] == t and count ==1:
            x = new_l[i][2]
            y = new_l[i][3]
            dist += ((float(x)-float(x_p))**2 + (float(y)-float(y_p))**2)**(1/2)
            x_p = x
            y_p = y
    if dist > max_dist:
        max_dist = dist
            
        
max_dist
    
```
