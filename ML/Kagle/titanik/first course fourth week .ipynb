## Импорт необходимых библиотек

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

## Распределение Mielke Beta-Kappa 

```python
k, s = 4, 5
mlk = sts.mielke(k,s)
theoretical_mean, theoretical_dispersion = mlk.stats(moments='mv')  
```

## Строим теоретический график распределения вероятности и экспериментальный

```python
sample = mlk.rvs(10000)
plt.hist(sample, bins = 40, density=True, label='histogram') # density=True нормализует гистограмму

x = np.linspace(0,4,100)
pdf = mlk.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')

plt.legend()
plt.ylabel('distribution density')
plt.xlabel('$x$')
```

## Вспомогательная функция создает матрицу количество выборок на количество сэмплов заполненную сэмплами из переданного распределения и применяет к строкам (массиву сэмплов) переданную функцию

```python
def func1d_sampler(numpy_func1d, distr_object, num_of_samples, num_of_selections):
    vector_of_samples = distr_object.rvs(num_of_samples*num_of_selections).reshape(num_of_selections, num_of_samples)
    return np.apply_along_axis(func1d=numpy_func1d,axis=1,arr=vector_of_samples)
```

## Строим теоретический график распределения вероятности средней и экспериментальный при количестве сэмплов n = 3, 5, 10, 30, 50, 100

```python
n=3
sample = func1d_sampler(np.average, mlk, n, 10000) # применяем функцию среднего к n сэмплам в 10 000 выборок
plt.hist(sample, bins = 40, density=True, label='histogram')

norm_rv = sts.norm(theoretical_mean.item(), np.sqrt(theoretical_dispersion.item()/n)) # нормальное распределение средних от n сэмплов
x = np.linspace(0.25,1.75,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pDF')

plt.legend()
plt.ylabel('distriburion density of means')
plt.xlabel('$x$')
```
